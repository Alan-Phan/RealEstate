```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('RealEstateData.csv')
df
```

```python
#Drop the first column since it is the index used in csv file
columns = list(df)
df.drop(columns[0],axis = 1, inplace=True)
df
```

The first column was an extra index that was created due to how the data was retrieved, it is deleted as it didn't serve a purpose.

```python
#since 1qtr bath is 0 for all rows delete that column as it provides no information
df.drop(columns=['baths_1qtr'],inplace=True)
```

No home in the dataset had a one-quarter bath so the column was also deleted.

```python
#No home has a name so we will delete that column as well
df.drop(columns=['name'], inplace=True)
```

Since there is only 0 in the name column we will delete that column as well.

Interestingly From this exploration I noticed that for the type column there was "land" which makes almost every other column 0 (except sold price) meaning that it may be better to delete these rows since my primary objective was to look primarily at house prices.

```python
#deleting rows that have "land" as their type
df.drop(df[df['type'] == 'land'].index, inplace=True)
df
```
